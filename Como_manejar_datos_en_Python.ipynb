## Listas

```python
# Construye una lista
mi_lista = ['A', 42, 'pupusa', 0.58]

print(mi_lista)
```

```python
# Añadir un elemento a una lista
mi_lista.append('Alberto')

print(mi_lista)
```

```python
# Eliminar un elemento de la lista
del mi_lista[1]

print(mi_lista)
```

```python
# Invertir el orden de la lista

mi_lista.reverse()

print(mi_lista)
```

```python
# Verificar si un objeto de Python es de hecho una lista

type(mi_lista)
```

Ejemplo de uso 1

```python
# Un equipo de marketing desea almacenar una lista de clientes que recibirán un correo electrónico
customers = ["Ana", "Luis", "Maria", "Alberto"]

# Llega un nuevo cliente
customers.append("Juan")

print(customers)
```

```python
# Una clienta se va

customers.remove("Maria")
print(customers)
```

Ejemplo de uso 2

```python
 # Una empresa quiere almacenar valores de ventas diarios para analizarlos

venta_diarias = [120, 340, 560, 230, 410]
print(venta_diarias)
```

```python
# Queremos saber la venta total y la mejor venta

venta_total = sum(venta_diarias)
mejor_venta = max(venta_diarias)

print(venta_total)
print(mejor_venta)
```

## Tuplas

```python

# Construye una lista
mi_tupla = ('Miguel', 32, 1.7)
print(mi_tupla)
```

```python
# Construye una lista (otra forma de hacer)
mi_tupla_2 = 'Maria', 30, 1.65
print(mi_tupla_2)
```

```python
# Tambien se puede crear a partir de una lista existente

mi_lista = ['A', 42, 'pupusa', 0.58]

mi_tupla_3 = tuple(mi_lista)
print(mi_tupla_3)
```

## Diccionarios

```python
# Construye un diccionario
mi_diccionario = {"nombre": "Alberto",
                  "edad": 32,
                  "ciudad": "San Salvador"}
print(mi_diccionario)
```

```python
# Otra forma de crear un diccionario
mi_diccionario = dict(nombre = "Alberto",
                  edad = 32,
                  ciudad = "San Salvador")
print(mi_diccionario)
```

```python
# Acceder a las clave, valor y pares clave-valor
mi_diccionario.keys()
mi_diccionario.values()
mi_diccionario.items()
```

```python
# Acceder al valor usando su clave
mi_diccionario['ciudad']
```

```python
mi_diccionario['edad']
```

```python
# Eliminar un elemento del diccionario
del(mi_diccionario['edad'])
print(mi_diccionario)
```

```python
# Cambiar un elemento en el diccionario
mi_diccionario = {"nombre": "Alberto",
                  "edad": 32,
                  "ciudad": "San Salvador"}

mi_diccionario['edad'] = 28
```

```python
# Agregar un elemento en el diccionario
mi_diccionario['estatura'] = 1.72
print(mi_diccionario)
```

Ejemplo de uso

```python
# Una tienda en línea necesita acceder rápidamente al estado de un pedido y poder actualizarlo cuando sea necesario
orders = {
    "ORD001": "Shipped",
    "ORD002": "Processing",
    "ORD003": "Delivered"
}

print(orders["ORD002"])
```

```python
# Actualizamos el valor pero la clave permanece igual
orders["ORD002"] = "Shipped"
print(orders)
```
